```python
import tensorflow as tf

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import cv2 as cv
```

```python
def get_class_name(code):
    labels = { 
    "apple" : 0, 
    "avocado" :1, 
    "banana" :2, 
    "cherry" :3, 
    "kiwi" :4, 
    "mango" :5, 
    "orange" :6, 
    "pinenapple" :7,
    "strawberries" :8, 
    "watermelon" :9 
}
    for x,y in labels.items():
        if code==y: return x
```

```python
get_class_name(1)
```

```python
def predict(img_path,model,img_shape):
    image = cv.imread(img_path)
    if image is None : return "img not found"
    image_array = [list(cv.resize(image , img_shape))]
    X = np.array(image_array)
    prediction = model.predict(X)
    class_name = get_class_name(np.argmax(prediction[0]))
    img = cv.cvtColor(cv.resize(image , img_shape), cv.COLOR_BGR2RGB)
    plt.subplot(1 , 1 , 1)
    plt.title(class_name)
    plt.imshow(img)
    plt.axis("off")
   
```

```python
model = tf.keras.models.load_model("./src/saved")
img_shape = (150,150)
```

```python
img_path = "./test_images/apples-1296x728-header.webp"
```

```python
predict(img_path,model,img_shape)
```
